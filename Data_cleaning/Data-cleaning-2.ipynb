## Handling missing values (imputing category data) how to fill missing values 

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(4) # to see first three rows    Nan show the  missing value 
```

```python
dataset.isnull().sum()
```

### we can fill the data by using forward filling , backward filling ,and using modes . write way to fill category data 

```python
dataset.info()
```

### this filling is done by using backward filling and rows wise 

```python
 dataset.fillna(method="bfill").head(5) 
```

### this filling is done by using forward filling and rows wise  but value will null in loan amount  bcz there is no data on upward 

```python
 dataset.fillna(method="ffill").head(5) 
```

### if i want my data is filling using column wise  backward or forward then we use "axis" if axis=0 means rows wise or if axis=1 mean column

```python
dataset.fillna(method="ffill",axis=1).head(5)
```

```python
 dataset.fillna(method="bfill",axis=1).head(5)
```

### using mode filling ,, the most repeatable data use techinique is mode 

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True) # this is done a particluar column 
```

```python
# if we filling all the object type data filling mode then we use this technique loop
for i in dataset.select_dtypes(include="object").columns:
    dataset[i].fillna(dataset[i].mode()[0], inplace=True)

```

```python
dataset.isnull().sum()
# the remaining data will be numerical all the category data will fill using mode 
```

 ## Ordinal encoding 

```python
# two methoid to perform ordinal encoding first is using sklearn and second is map function
import pandas as pd 
```

```python
# we create a dataset 
df = pd.DataFrame({"size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head(3)
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["size"]])
```

```python
 df["size_en"]=oe.transform(df[["size"]])
```

```python
df
```

### second method to perform ordinal using map function 


```python
# first we create a dictionary 
ord_data1 = {"s":0,"m":1,"l":2,"xl":3}  # these number will chose randomly  any number like 5,6,7,8 or any thing 
```

```python
df["size_en-map"] =df["size"].map(ord_data1)
```

```python
df
```

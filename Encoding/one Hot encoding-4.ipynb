### Encoding mean when we have categorical data then we convert inti numerical data that process is encoding


```python
import pandas as pd
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(4) # to see first three rows    Nan show the  missing value 
```

```python

dataset.isnull().sum()
```

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python
dataset["Married"].fillna(dataset["Married"].mode()[0],inplace=True)
```

```python
en_data = dataset[["Gender","Married"]]
en_data
```

```python
dataset.isnull().sum()
```

### performing encoding using the "get_dumies" function that is present in the pandas


```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
# this is boolen so we use sklearn for encoding 
```

```python
# from sklearn library perform encoding using the "onehotencoder " function 
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(en_data)
```

```python
ohe = OneHotEncoder()
ar= ohe.fit_transform(en_data).toarray() # this is a data in the form of array but  we need in column so we convert in to it 
ar
```

```python
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male"," Married_No"," Married_Yes"])
```
